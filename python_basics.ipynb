# Python Basics Notebook

This notebook introduces 9th and 10th graders to the basics of Python.

```python
# 1. Introduction to Python
print("Hello, world!")  # This prints a message to the screen
```

```python
# 2. Variables & Data Types
name = "Alice"  # String
age = 14  # Integer
height = 5.4  # Float
is_student = True  # Boolean

# Displaying variable values
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is student?", is_student)
```

```python
# 3. Basic Operations
sum_result = 5 + 3  # Addition
product = 4 * 2  # Multiplication
message = "Python is " + "fun!"  # String concatenation
print(sum_result, product, message)
```

```python
# 4. Control Flow
x = 10
if x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or less")
```

```python
# 5. Loops
for i in range(5):
    print("Loop iteration:", i)
```

```python
# 6. Functions
def greet(name):
    return "Hello, " + name + "!"

print(greet("Alice"))
```

```python
# 7. Lists & Dictionaries
fruits = ["apple", "banana", "cherry"]
print("First fruit:", fruits[0])

student = {"name": "Alice", "age": 14}
print("Student Name:", student["name"])
```
